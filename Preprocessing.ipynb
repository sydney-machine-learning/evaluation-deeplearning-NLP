```python
file = open('Data/ArthurWaley.txt', 'r')
lines = file.readlines()

lines = [l.strip() for l in lines if (l.strip() != '' and l.strip() != 'up')]


for i in range(len(lines)):
    if lines[i].isdigit():
        chapter = lines[i]
        start_number = i
        end_number = len(lines)
        for j in range(i+1, len(lines)):
            if lines[j].isdigit():
                end_number = j
                break

        for x in range(start_number+1, end_number):
            lines[x] = lines[x]+ f'|{chapter}\n'

lines = [l for l in lines if (l.isdigit() == False)]
print(lines)
lines.insert(0, 'Text|Chapter\n')
print(lines)
```

```python

with open('Data/ArthurWaley.csv', 'w') as f:
    f.writelines(lines)
```

```python
file = open('Data/LinYutang.txt', 'r')
lines = file.readlines()

lines = [l.strip() for l in lines if (l.strip() != '' )]

print(lines[0:10])

for i in range(len(lines)):
    if lines[i].strip()[0].isdigit():
        chapter = lines[i].split('.')[0]
        print(chapter)
        start_number = i
        end_number = len(lines)
        for j in range(i+1, len(lines)):
            if lines[j].strip()[0].isdigit():
                end_number = j
                break

        for x in range(start_number+1, end_number):
            lines[x] = lines[x]+ f'|{chapter}\n'

lines = [l for l in lines if not l.strip()[0].isdigit()]
#print(lines)
lines.insert(0, 'Text|Chapter\n')
print(lines)
```

```python
with open('Data/LinYutang.csv', 'w') as f:
    f.writelines(lines)
```
